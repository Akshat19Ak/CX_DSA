```python
# 1886. Determine Whether Matrix Can Be Obtained By Rotation 
from typing import List

class Solution:
    def findRotation(self, mat: List[List[int]], target: List[List[int]]) -> bool:
        
        # Function to rotate matrix 90° clockwise IN-PLACE
        def rotate(matrix):
            n = len(matrix)
            
            # Step 1: Transpose (swap rows and columns)
            for i in range(n):
                for j in range(i, n):
                    matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
            
            # Step 2: Reverse each row
            for row in matrix:
                row.reverse()
        
        # Try all 4 possible rotations
        for _ in range(4):
            if mat == target:   # Check if matrices match
                return True
            rotate(mat)         # Rotate in-place
        
        return False


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem checks whether we can convert one matrix into another by rotating it.
# Since a square matrix has only 4 possible orientations (0°, 90°, 180°, 270°),
# we simply try all rotations and check for equality. The brute force approach
# creates a new matrix each time, while the optimal approach uses an in-place
# rotation technique (transpose + reverse rows), which reduces space complexity.
# This guarantees all test cases are handled efficiently.

# IMPORTANT INSIGHTS:
# 1. Only 4 rotations exist → no need for complex algorithms
# 2. Rotation can be done using transpose + reverse
# 3. Matrix comparison is straightforward using ==

# PATTERN USED:
# Matrix Transformation Pattern → (Transpose + Reverse)

# WHERE ELSE THIS PATTERN IS USED:
# 1. Rotate Image (LeetCode 48)
# 2. Image processing / Computer Vision
# 3. Grid transformation problems
# 4. Any problem involving 2D matrix rotation

# NOTE:
# Brute Force → Time: O(n²), Space: O(n²)
# Optimal     → Time: O(n²), Space: O(1)

# --- ADD THIS TO DRY RUN ---
sol = Solution()
matrix = [[0, 1], [1, 1]]
target = [[1, 1], [1, 0]]
print(sol.findRotation(matrix, target))
```

```python
# LeetCode 3546: Equal Sum Grid Partition I

from typing import List

class Solution:
    def canPartitionGrid(self, grid: List[List[int]]) -> bool:
        
        m, n = len(grid), len(grid[0])
        
        # Step 1: Compute total sum of grid
        total_sum = 0
        for row in grid:
            total_sum += sum(row)
        
        # If total sum is odd → cannot split equally
        if total_sum % 2 != 0:
            return False
        
        target = total_sum // 2
        
        # Step 2: Try horizontal cuts
        prefix_sum = 0
        for i in range(m - 1):  # ensure non-empty bottom
            prefix_sum += sum(grid[i])
            if prefix_sum == target:
                return True
        
        # Step 3: Try vertical cuts
        prefix_sum = 0
        for j in range(n - 1):  # ensure non-empty right
            col_sum = 0
            for i in range(m):
                col_sum += grid[i][j]
            prefix_sum += col_sum
            
            if prefix_sum == target:
                return True
        
        return False


# ----------- TEST CASE (for practice) -----------
# grid = [[1,4],[2,3]]
# obj = Solution()
# print(obj.canPartitionGrid(grid))   # Expected: True


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem asks whether we can divide a grid into two non-empty parts
# using a horizontal or vertical cut such that both parts have equal sum.
# The brute force approach checks all cuts and recomputes sums repeatedly,
# which is inefficient. The optimal approach uses prefix sum logic:
# first compute total sum, then iterate through rows or columns while
# maintaining a running sum and check if it equals half of total.
# This ensures all possible valid cuts are checked efficiently.

# NOTE:
# Brute Force → Time: O((m+n) × m × n), Space: O(1)
# Optimal     → Time: O(m × n), Space: O(1)

# IMPORTANT INSIGHTS:
# 1. Equal partition → total sum must be even
# 2. Only need to find prefix sum = total/2
# 3. Ensure both partitions are non-empty

# PATTERN USED:
# Prefix Sum + Partition Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. Partition Equal Subset Sum
# 2. Split Array problems
# 3. Subarray with given sum
# 4. Load balancing / workload distribution problems
```
