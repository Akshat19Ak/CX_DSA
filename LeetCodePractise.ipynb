```python
# 1886. Determine Whether Matrix Can Be Obtained By Rotation 
from typing import List

class Solution:
    def findRotation(self, mat: List[List[int]], target: List[List[int]]) -> bool:
        
        # Function to rotate matrix 90° clockwise IN-PLACE
        def rotate(matrix):
            n = len(matrix)
            
            # Step 1: Transpose (swap rows and columns)
            for i in range(n):
                for j in range(i, n):
                    matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
            
            # Step 2: Reverse each row
            for row in matrix:
                row.reverse()
        
        # Try all 4 possible rotations
        for _ in range(4):
            if mat == target:   # Check if matrices match
                return True
            rotate(mat)         # Rotate in-place
        
        return False


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem checks whether we can convert one matrix into another by rotating it.
# Since a square matrix has only 4 possible orientations (0°, 90°, 180°, 270°),
# we simply try all rotations and check for equality. The brute force approach
# creates a new matrix each time, while the optimal approach uses an in-place
# rotation technique (transpose + reverse rows), which reduces space complexity.
# This guarantees all test cases are handled efficiently.

# IMPORTANT INSIGHTS:
# 1. Only 4 rotations exist → no need for complex algorithms
# 2. Rotation can be done using transpose + reverse
# 3. Matrix comparison is straightforward using ==

# PATTERN USED:
# Matrix Transformation Pattern → (Transpose + Reverse)

# WHERE ELSE THIS PATTERN IS USED:
# 1. Rotate Image (LeetCode 48)
# 2. Image processing / Computer Vision
# 3. Grid transformation problems
# 4. Any problem involving 2D matrix rotation

# NOTE:
# Brute Force → Time: O(n²), Space: O(n²)
# Optimal     → Time: O(n²), Space: O(1)

# --- ADD THIS TO DRY RUN ---
sol = Solution()
matrix = [[0, 1], [1, 1]]
target = [[1, 1], [1, 0]]
print(sol.findRotation(matrix, target))
```

```python
# LeetCode 3546: Equal Sum Grid Partition I

from typing import List

class Solution:
    def canPartitionGrid(self, grid: List[List[int]]) -> bool:
        
        m, n = len(grid), len(grid[0])
        
        # Step 1: Compute total sum of grid
        total_sum = 0
        for row in grid:
            total_sum += sum(row)
        
        # If total sum is odd → cannot split equally
        if total_sum % 2 != 0:
            return False
        
        target = total_sum // 2
        
        # Step 2: Try horizontal cuts
        prefix_sum = 0
        for i in range(m - 1):  # ensure non-empty bottom
            prefix_sum += sum(grid[i])
            if prefix_sum == target:
                return True
        
        # Step 3: Try vertical cuts
        prefix_sum = 0
        for j in range(n - 1):  # ensure non-empty right
            col_sum = 0
            for i in range(m):
                col_sum += grid[i][j]
            prefix_sum += col_sum
            
            if prefix_sum == target:
                return True
        
        return False


# ----------- TEST CASE (for practice) -----------
# grid = [[1,4],[2,3]]
# obj = Solution()
# print(obj.canPartitionGrid(grid))   # Expected: True


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem asks whether we can divide a grid into two non-empty parts
# using a horizontal or vertical cut such that both parts have equal sum.
# The brute force approach checks all cuts and recomputes sums repeatedly,
# which is inefficient. The optimal approach uses prefix sum logic:
# first compute total sum, then iterate through rows or columns while
# maintaining a running sum and check if it equals half of total.
# This ensures all possible valid cuts are checked efficiently.

# NOTE:
# Brute Force → Time: O((m+n) × m × n), Space: O(1)
# Optimal     → Time: O(m × n), Space: O(1)

# IMPORTANT INSIGHTS:
# 1. Equal partition → total sum must be even
# 2. Only need to find prefix sum = total/2
# 3. Ensure both partitions are non-empty

# PATTERN USED:
# Prefix Sum + Partition Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. Partition Equal Subset Sum
# 2. Split Array problems
# 3. Subarray with given sum
# 4. Load balancing / workload distribution problems
```

```python
# LeetCode 1: Two Sum

from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        
        # Dictionary to store value -> index
        seen = {}
        
        # Iterate through array
        for i, num in enumerate(nums):
            
            # Find required complement
            complement = target - num
            
            # If complement exists, return indices
            if complement in seen:
                return [seen[complement], i]
            
            # Otherwise store current number with index
            seen[num] = i
        
        return []  # not needed as solution always exists


# ----------- TEST CASE (for practice) -----------
# nums = [3,2,4]
# target = 6
# obj = Solution()
# print(obj.twoSum(nums, target))  # Expected: [1,2]


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem asks us to find two indices in an array such that their values add up to a target.
# The brute force approach checks all pairs, but it is slow (O(n²)).
# The optimal solution uses a hashmap to store previously seen numbers and checks if the
# complement (target - current number) exists. This reduces time complexity to O(n).
# Since exactly one solution exists, we can return immediately when found.

# NOTE:
# Brute Force → Time: O(n²), Space: O(1)
# Optimal     → Time: O(n), Space: O(n)

# IMPORTANT INSIGHTS:
# 1. Instead of finding two numbers, find complement
# 2. HashMap gives O(1) lookup
# 3. One-pass solution is enough

# PATTERN USED:
# HashMap + Complement Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. 3Sum / 4Sum problems
# 2. Subarray Sum Equals K
# 3. Two Sum II (sorted array)
# 4. Pair with given difference problems
```

```python
# LeetCode 2573: Find the String with LCP

from typing import List

class Solution:
    def findTheString(self, lcp: List[List[int]]) -> str:
        
        n = len(lcp)
        res = [''] * n
        
        # Step 1: Assign characters greedily
        ch = ord('a')
        
        for i in range(n):
            if res[i] == '':
                if ch > ord('z'):  # cannot assign more than 26 chars
                    return ""
                
                # Assign current character
                res[i] = chr(ch)
                
                # Propagate to all j where lcp[i][j] > 0
                for j in range(i + 1, n):
                    if lcp[i][j] > 0:
                        res[j] = res[i]
                
                ch += 1
        
        # Step 2: Validate LCP matrix
        # Build computed LCP
        computed = [[0]*n for _ in range(n)]
        
        for i in range(n-1, -1, -1):
            for j in range(n-1, -1, -1):
                if res[i] == res[j]:
                    if i == n-1 or j == n-1:
                        computed[i][j] = 1
                    else:
                        computed[i][j] = computed[i+1][j+1] + 1
        
        # Compare matrices
        for i in range(n):
            for j in range(n):
                if computed[i][j] != lcp[i][j]:
                    return ""
        
        return "".join(res)


# ----------- TEST CASE (for practice) -----------
# lcp = [[4,0,2,0],[0,3,0,1],[2,0,2,0],[0,1,0,1]]
# obj = Solution()
# print(obj.findTheString(lcp))   # Expected: "abab"


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem asks us to reconstruct a string from its LCP matrix.
# Brute force is impossible because of exponential possibilities.
# The optimal approach uses greedy assignment: if lcp[i][j] > 0,
# then characters at i and j must be equal. We assign smallest possible
# characters to maintain lexicographically smallest string.
# Finally, we validate the constructed string by recomputing LCP.
# If mismatch occurs, return empty string.

# NOTE:
# Brute Force → Time: Exponential, Space: O(n)
# Optimal     → Time: O(n²), Space: O(n)

# IMPORTANT INSIGHTS:
# 1. lcp[i][j] > 0 ⇒ characters must be same
# 2. Use greedy to ensure lexicographically smallest string
# 3. Always validate result

# PATTERN USED:
# Greedy + Constraint Propagation + Validation

# WHERE ELSE THIS PATTERN IS USED:
# 1. Graph coloring problems
# 2. Union-Find grouping problems
# 3. String reconstruction problems
# 4. Constraint satisfaction problems
```

```python
# LeetCode 2840: Check if Strings Can be Made Equal With Operations II

class Solution:
    def checkStrings(self, s1: str, s2: str) -> bool:
        
        n = len(s1)
        
        # Frequency arrays for even and odd indices
        even_count = [0] * 26
        odd_count = [0] * 26
        
        # Count for s1
        for i in range(n):
            if i % 2 == 0:
                even_count[ord(s1[i]) - ord('a')] += 1
                even_count[ord(s2[i]) - ord('a')] -= 1
            else:
                odd_count[ord(s1[i]) - ord('a')] += 1
                odd_count[ord(s2[i]) - ord('a')] -= 1
        
        # Check if all counts match
        return all(x == 0 for x in even_count) and all(x == 0 for x in odd_count)


# ----------- TEST CASE (for practice) -----------
# s1 = "abcdba"
# s2 = "cabdab"
# obj = Solution()
# print(obj.canBeEqual(s1, s2))  # Expected: True


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem allows swapping characters at positions where index difference is even.
# This means characters can only move within even indices or odd indices.
# So instead of simulating swaps, we group indices into even and odd positions,
# and compare frequency of characters in both groups for s1 and s2.
# If both groups match, transformation is possible.

# NOTE:
# Brute Force → Time: Exponential, Space: O(n)
# Optimal     → Time: O(n), Space: O(1)

# IMPORTANT INSIGHTS:
# 1. Even difference → same parity indices
# 2. Groups are independent
# 3. Compare frequencies instead of simulating swaps

# PATTERN USED:
# Grouping + Frequency Counting + Invariant Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. LeetCode 2839 (Operations I)
# 2. Parity-based grouping problems
# 3. Graph connected components (restricted swaps)
# 4. String transformation problems
```

```python
# LeetCode 9: Palindrome Number

class Solution:
    def isPalindrome(self, x: int) -> bool:
        
        # Negative numbers are not palindrome
        if x < 0:
            return False
        
        # Numbers ending with 0 (except 0 itself) are not palindrome
        if x % 10 == 0 and x != 0:
            return False
        
        reversed_half = 0
        
        # Reverse half of the number
        while x > reversed_half:
            digit = x % 10                     # Extract last digit
            reversed_half = reversed_half * 10 + digit
            x //= 10                          # Remove last digit
        
        # For even digits: x == reversed_half
        # For odd digits: ignore middle digit using reversed_half // 10
        return x == reversed_half or x == reversed_half // 10


# ----------- TEST CASE (for practice) -----------
# x = 121
# obj = Solution()
# print(obj.isPalindrome(x))  # Expected: True


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem asks us to check whether a number reads the same forward and backward.
# A brute force approach converts the number to string and compares with reverse.
# The optimal solution avoids string conversion and instead reverses only half of
# the number, then compares both halves. This avoids unnecessary work and prevents overflow.
# This approach ensures efficient and correct handling of all cases.

# NOTE:
# Brute Force → Time: O(n), Space: O(n)
# Optimal     → Time: O(log n), Space: O(1)

# IMPORTANT INSIGHTS:
# 1. Negative numbers are not palindrome
# 2. Only need to reverse half of number
# 3. Handle odd and even length separately

# PATTERN USED:
# Digit Manipulation + Half Reversal Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. Reverse Integer problem
# 2. Number digit problems
# 3. Math-based palindrome checks
# 4. Overflow-safe number operations
```

```python
# LeetCode 2751: Robot Collisions

from typing import List

class Solution:
    def survivedRobotsHealths(self, positions: List[int], healths: List[int], directions: str) -> List[int]:
        
        n = len(positions)
        
        # Combine robot data: position, health, direction, original index
        robots = [(positions[i], healths[i], directions[i], i) for i in range(n)]
        
        # Sort robots by position
        robots.sort()
        
        stack = []  # will store indices of robots moving right
        alive = [True] * n
        
        # Convert to list for mutation
        healths = healths[:]
        
        for pos, h, d, idx in robots:
            
            if d == 'R':
                stack.append(idx)
            else:
                # Handle collisions
                while stack and healths[idx] > 0:
                    top = stack[-1]
                    
                    if healths[top] < healths[idx]:
                        # Right robot dies
                        alive[top] = False
                        stack.pop()
                        healths[idx] -= 1
                    
                    elif healths[top] > healths[idx]:
                        # Left robot dies
                        alive[idx] = False
                        healths[top] -= 1
                        break
                    
                    else:
                        # Both die
                        alive[top] = False
                        alive[idx] = False
                        stack.pop()
                        break
        
        # Collect survivors in original order
        result = []
        for i in range(n):
            if alive[i]:
                result.append(healths[i])
        
        return result


# ----------- TEST CASE (for practice) -----------
# positions = [3,5,2,6]
# healths = [10,10,15,12]
# directions = "RLRL"
# obj = Solution()
# print(obj.survivedRobotsHealths(positions, healths, directions))  # Expected: [14]


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem simulates robots moving and colliding on a line.
# Only collisions between a right-moving robot and a left-moving robot matter.
# A brute force simulation is inefficient. The optimal solution sorts robots by position
# and uses a stack to track right-moving robots. When a left-moving robot is encountered,
# we simulate collisions using the stack until no more collisions are possible.
# This ensures efficient handling of all interactions.

# NOTE:
# Brute Force → Time: O(n²), Space: O(n)
# Optimal     → Time: O(n log n), Space: O(n)

# IMPORTANT INSIGHTS:
# 1. Only R vs L collisions matter
# 2. Sorting simplifies collision order
# 3. Stack helps track active robots

# PATTERN USED:
# Stack + Collision Simulation Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. Asteroid Collision (LeetCode 735)
# 2. Parentheses matching problems
# 3. Next Greater Element
# 4. Stack-based simulation problems
```

```python
# LeetCode 20: Valid Parentheses

class Solution:
    def isValid(self, s: str) -> bool:
        
        # Stack to store opening brackets
        stack = []
        
        # Mapping closing → opening
        mapping = {
            ')': '(',
            ']': '[',
            '}': '{'
        }
        
        # Traverse string
        for ch in s:
            
            # If opening bracket, push to stack
            if ch in mapping.values():
                stack.append(ch)
            
            else:
                # If stack is empty or mismatch
                if not stack or stack[-1] != mapping[ch]:
                    return False
                
                # Pop matched opening bracket
                stack.pop()
        
        # If stack is empty → valid
        return len(stack) == 0


# ----------- TEST CASE (for practice) -----------
# s = "([])"
# obj = Solution()
# print(obj.isValid(s))  # Expected: True


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem checks whether a string of brackets is valid.
# A brute force approach removes valid pairs repeatedly, but it is inefficient.
# The optimal solution uses a stack to track opening brackets.
# When a closing bracket appears, we check if it matches the top of the stack.
# If mismatch occurs or stack is not empty at the end, the string is invalid.

# NOTE:
# Brute Force → Time: O(n²), Space: O(n)
# Optimal     → Time: O(n), Space: O(n)

# IMPORTANT INSIGHTS:
# 1. Stack ensures correct order (LIFO)
# 2. Mapping helps match brackets
# 3. Final stack must be empty

# PATTERN USED:
# Stack + Matching Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. Asteroid Collision
# 2. Next Greater Element
# 3. Expression evaluation
# 4. XML/HTML tag validation
```

```python
# LeetCode 3418: Maximum Amount of Money Robot Can Earn

from typing import List

class Solution:
    def maximumAmount(self, coins: List[List[int]]) -> int:
        
        m, n = len(coins), len(coins[0])
        
        # dp[i][j][k] → max coins at (i,j) using k neutralizations
        dp = [[[-float('inf')] * 3 for _ in range(n)] for _ in range(m)]
        
        # Base case
        for k in range(3):
            dp[0][0][k] = coins[0][0] if coins[0][0] >= 0 else 0 if k > 0 else coins[0][0]
        
        for i in range(m):
            for j in range(n):
                if i == 0 and j == 0:
                    continue
                
                for k in range(3):
                    
                    best_prev = -float('inf')
                    
                    if i > 0:
                        best_prev = max(best_prev, dp[i-1][j][k])
                    if j > 0:
                        best_prev = max(best_prev, dp[i][j-1][k])
                    
                    if coins[i][j] >= 0:
                        dp[i][j][k] = best_prev + coins[i][j]
                    else:
                        # Option 1: take loss
                        dp[i][j][k] = best_prev + coins[i][j]
                        
                        # Option 2: neutralize
                        if k > 0:
                            best_prev_neutral = -float('inf')
                            if i > 0:
                                best_prev_neutral = max(best_prev_neutral, dp[i-1][j][k-1])
                            if j > 0:
                                best_prev_neutral = max(best_prev_neutral, dp[i][j-1][k-1])
                            
                            dp[i][j][k] = max(dp[i][j][k], best_prev_neutral)
        
        return max(dp[m-1][n-1])


# ----------- TEST CASE (for practice) -----------
# coins = [[0,1,-1],[1,-2,3],[2,-3,4]]
# obj = Solution()
# print(obj.maximumAmount(coins))  # Expected: 8


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem is a grid DP problem where we want maximum sum path,
# but with a twist: we can neutralize up to 2 negative cells.
# Brute force is exponential, so we use DP with state:
# (i, j, k) where k = number of neutralizations used.
# At each step, we choose whether to neutralize a negative cell or not.
# This ensures we explore all optimal paths efficiently.

# NOTE:
# Brute Force → Time: Exponential, Space: O(path)
# Optimal     → Time: O(m*n), Space: O(m*n)

# IMPORTANT INSIGHTS:
# 1. Extra state needed for neutralizations
# 2. Negative cells give choice → use or skip
# 3. Classic grid DP with twist

# PATTERN USED:
# Grid DP + Multi-State DP

# WHERE ELSE THIS PATTERN IS USED:
# 1. Cherry Pickup problems
# 2. Grid path with constraints
# 3. Knapsack variations
# 4. DP with limited operations
```

```python
# LeetCode 13: Roman to Integer

class Solution:
    def romanToInt(self, s: str) -> int:
        
        # Mapping of Roman numerals to integer values
        values = {
            'I': 1, 'V': 5, 'X': 10,
            'L': 50, 'C': 100,
            'D': 500, 'M': 1000
        }
        
        total = 0
        
        # Traverse string
        for i in range(len(s)):
            
            # If current value is less than next → subtract
            if i < len(s) - 1 and values[s[i]] < values[s[i + 1]]:
                total -= values[s[i]]
            else:
                total += values[s[i]]
        
        return total


# ----------- TEST CASE (for practice) -----------
# s = "MCMXCIV"
# obj = Solution()
# print(obj.romanToInt(s))  # Expected: 1994


# ===================== NOTES FOR REVISION =====================

# BEGINNER-FRIENDLY INTRODUCTION:
# This problem converts a Roman numeral into an integer.
# Normally, values are added from left to right, but if a smaller value appears
# before a larger one, we subtract it (like IV = 4).
# The optimal solution uses a greedy approach: compare each character with the next.
# If current is smaller than next → subtract, otherwise → add.
# This handles all cases efficiently in one pass.

# NOTE:
# Brute Force → Time: O(n), Space: O(1)
# Optimal     → Time: O(n), Space: O(1)

# IMPORTANT INSIGHTS:
# 1. Subtraction occurs when smaller number is before larger
# 2. Lookahead comparison simplifies logic
# 3. Single pass solution is enough

# PATTERN USED:
# Greedy + Lookahead Comparison Pattern

# WHERE ELSE THIS PATTERN IS USED:
# 1. String parsing problems
# 2. Expression evaluation
# 3. Next greater element type problems
# 4. Greedy decision problems
```

```python
#qn 3661 hard one
from typing import List
import bisect

class Solution:
    def maxWalls(self, robots: List[int], distance: List[int], walls: List[int]) -> int:
        walls_sorted = sorted(walls)
        robot_pos_set = set(robots)

        def count_walls(lo, hi):
            if lo > hi:
                return 0
            return bisect.bisect_right(walls_sorted, hi) - bisect.bisect_left(walls_sorted, lo)

        n = len(robots)
        robots_data = sorted(zip(robots, distance))
        r = [pos for pos, _ in robots_data]
        d = [dist for _, dist in robots_data]

        # Walls at robot positions are always destroyed by that robot
        walls_at_robots = sum(1 for w in walls if w in robot_pos_set)

        # dp[0] = max walls if current robot fires LEFT
        # dp[1] = max walls if current robot fires RIGHT
        left_tail = count_walls(r[0] - d[0], r[0] - 1)
        dp = [left_tail, 0]

        for i in range(1, n):
            # Strict gap: (r[i-1], r[i]) exclusive of robot positions
            R_lo, R_hi = r[i-1] + 1, min(r[i-1] + d[i-1], r[i] - 1)
            L_lo, L_hi = max(r[i] - d[i], r[i-1] + 1), r[i] - 1

            total_R = count_walls(R_lo, R_hi)       # prev fires right into gap
            total_L = count_walls(L_lo, L_hi)       # curr fires left into gap
            B       = count_walls(max(R_lo, L_lo), min(R_hi, L_hi))  # overlap

            # ✅ Fixed transitions:
            # prev=LEFT,  curr=LEFT  → gain total_L   (not just C!)
            # prev=RIGHT, curr=LEFT  → gain total_R + total_L - B (union)
            # prev=LEFT,  curr=RIGHT → gain 0
            # prev=RIGHT, curr=RIGHT → gain total_R
            new_L = max(dp[0] + total_L,
                        dp[1] + total_R + total_L - B)
            new_R = max(dp[0] + 0,
                        dp[1] + total_R)
            dp = [new_L, new_R]

        right_tail = count_walls(r[-1] + 1, r[-1] + d[-1])
        return max(dp[0], dp[1] + right_tail) + walls_at_robots
```

```python
# Question: Decode the Slanted Ciphertext (LeetCode 2075)
# Goal: Recover original text from diagonally encoded matrix

class Solution:
    def decodeCiphertext(self, encodedText: str, rows: int) -> str:
        
        # Edge Case:
        # If there is only 1 row, encoding didn't change anything
        if rows == 1:
            return encodedText
        
        # Total number of characters in encoded string
        total_length = len(encodedText)
        
        # Calculate number of columns in matrix
        cols = total_length // rows
        
        # This will store our decoded characters
        result = []
        
        # Start diagonal traversal from each column of first row
        for start_col in range(cols):
            
            # Start from (row=0, col=start_col)
            r = 0
            c = start_col
            
            # Traverse diagonally (down-right direction)
            while r < rows and c < cols:
                
                # Convert 2D index → 1D index
                index = r * cols + c
                
                # Add character to result
                result.append(encodedText[index])
                
                # Move diagonally
                r += 1
                c += 1
        
        # Join characters and remove trailing spaces
        return ''.join(result).rstrip()


"""
🔁 ALTERNATIVE METHOD (Matrix-Based - Less Optimal)

# Uncomment this method if you want to understand matrix simulation

class Solution:
    def decodeCiphertext(self, encodedText: str, rows: int) -> str:
        if rows == 1:
            return encodedText
        
        n = len(encodedText)
        cols = n // rows
        
        # Build matrix
        matrix = []
        idx = 0
        
        for r in range(rows):
            row = []
            for c in range(cols):
                row.append(encodedText[idx])
                idx += 1
            matrix.append(row)
        
        result = []
        
        # Diagonal traversal
        for start_col in range(cols):
            r, c = 0, start_col
            while r < rows and c < cols:
                result.append(matrix[r][c])
                r += 1
                c += 1
        
        return ''.join(result).rstrip()
"""
```

```python
# Question: Merge two sorted arrays into nums1 in-place

from typing import List

class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        
        # Pointer for nums1's last valid element
        i = m - 1
        
        # Pointer for nums2's last element
        j = n - 1
        
        # Pointer for final position in nums1
        k = m + n - 1
        
        # Traverse from the back
        while i >= 0 and j >= 0:
            
            # Compare elements from nums1 and nums2
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]  # place larger element
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1
            
            k -= 1  # move result pointer
        
        # If nums2 still has elements left, copy them
        while j >= 0:
            nums1[k] = nums2[j]
            j -= 1
            k -= 1


"""
🔁 ALTERNATIVE METHOD 1 (Sorting - Not Optimal)

# nums1[:m] = nums1[:m] + nums2
# nums1.sort()

Time: O((m+n) log(m+n))


🔁 ALTERNATIVE METHOD 2 (Extra Array)

# temp = []
# i, j = 0, 0

# while i < m and j < n:
#     if nums1[i] < nums2[j]:
#         temp.append(nums1[i])
#         i += 1
#     else:
#         temp.append(nums2[j])
#         j += 1

# temp.extend(nums1[i:m])
# temp.extend(nums2[j:])

# nums1[:] = temp

Time: O(m+n), Space: O(m+n)
"""
```

```python
# Question:3653 XOR After Range Multiplication Queries I

from typing import List

class Solution:
    def xorAfterQueries(self, nums: List[int], queries: List[List[int]]) -> int:
        
        MOD = 10**9 + 7  # given modulo
        
        # Process each query
        for li, ri, ki, vi in queries:
            
            # Start from li and jump with step ki
            idx = li
            
            while idx <= ri:
                
                # Multiply and take modulo
                nums[idx] = (nums[idx] * vi) % MOD
                
                # Jump to next valid index
                idx += ki
        
        # Compute XOR of all elements
        result = 0
        
        for num in nums:
            result ^= num
        
        return result


"""
🔁 ALTERNATIVE METHOD (Using for-loop instead of while)

for li, ri, ki, vi in queries:
    for idx in range(li, ri + 1, ki):
        nums[idx] = (nums[idx] * vi) % MOD
"""
```

```python
# Question: Given numRows, return first numRows of Pascal's Triangle
# Example: numRows = 5 → [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]

# PURPOSE:
# Generate Pascal's Triangle where each element is sum of two elements above it

# APPROACH (Step-by-step flow):
# 1. Initialize result list to store all rows
# 2. Loop from row = 0 to numRows-1
# 3. For each row:
#    - Create a new row list
#    - First and last elements = 1
#    - For middle elements:
#         current[j] = previous[j-1] + previous[j]
# 4. Append row to result
# 5. Return result

# ASSUMPTIONS:
# - numRows ≥ 1
# - Output must contain exactly numRows rows


from typing import List

class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        
        # Final result list that will store all rows
        triangle = []
        
        # Loop through each row index
        for row in range(numRows):
            
            # Create a new row
            current_row = []
            
            # Loop through each column in this row
            for col in range(row + 1):
                
                # CASE 1: First or Last element of row
                # Always 1 (boundary condition)
                if col == 0 or col == row:
                    current_row.append(1)
                
                # CASE 2: Middle elements
                else:
                    # Value comes from previous row
                    # Sum of two elements above
                    value = triangle[row - 1][col - 1] + triangle[row - 1][col]
                    current_row.append(value)
            
            # Add current row to triangle
            triangle.append(current_row)
        
        # Return final Pascal triangle
        return triangle


"""
🔁 ALTERNATIVE METHOD 1 (Using nCr formula)

# Formula: nCr = n! / (r! * (n-r)!)
# Each row can be generated using combinations

def generate(self, numRows):
    from math import comb
    res = []
    for i in range(numRows):
        row = []
        for j in range(i+1):
            row.append(comb(i, j))
        res.append(row)
    return res

Time: O(n^3) due to factorials


🔁 ALTERNATIVE METHOD 2 (Optimized nCr iterative)

# Using relation:
# C(n, r) = C(n, r-1) * (n-r+1) / r

def generate(self, numRows):
    res = []
    for i in range(numRows):
        row = [1]
        val = 1
        for j in range(1, i+1):
            val = val * (i - j + 1) // j
            row.append(val)
        res.append(row)
    return res

Time: O(n^2)
"""
```

```python
# Question136: Find the element that appears only once (others appear twice)

# PURPOSE:
# Identify the unique element using XOR properties in O(n) time and O(1) space

# APPROACH (Step-by-step flow):
# 1. Initialize a variable result = 0
# 2. Traverse the array
# 3. XOR each element with result
# 4. Duplicate elements cancel out (a ^ a = 0)
# 5. Final result will be the single unique element

# ASSUMPTIONS:
# - Exactly one element appears once
# - All others appear exactly twice


from typing import List

class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        
        # Initialize result with 0
        # Why 0? Because XOR with 0 gives the number itself
        result = 0
        
        # Traverse all elements in the array
        for num in nums:
            
            # Apply XOR operation
            # This cancels duplicates automatically
            result ^= num
        
        # After processing all elements,
        # result will contain the unique number
        return result


"""
🔁 ALTERNATIVE METHOD 1 (Using HashMap)

def singleNumber(self, nums):
    freq = {}
    
    for num in nums:
        freq[num] = freq.get(num, 0) + 1
    
    for key in freq:
        if freq[key] == 1:
            return key

Time: O(n), Space: O(n)


🔁 ALTERNATIVE METHOD 2 (Using Set Trick)

def singleNumber(self, nums):
    return 2 * sum(set(nums)) - sum(nums)

Time: O(n), Space: O(n)
"""
```

```python
# Question:3488. Closest Equal Element Queries 
# Correct and clean circular handling

from collections import defaultdict
from typing import List

class Solution:
    def solveQueries(self, nums: List[int], queries: List[int]) -> List[int]:
        
        n = len(nums)
        
        # Step 1: Store indices of each value
        pos = defaultdict(list)
        for i, num in enumerate(nums):
            pos[num].append(i)
        
        # Step 2: Prepare answer array
        ans = [-1] * n
        
        # Step 3: Process each value group
        for indices in pos.values():
            
            # If only one occurrence → no valid answer
            if len(indices) == 1:
                continue
            
            m = len(indices)
            
            for i in range(m):
                curr = indices[i]
                
                # Previous index (circular)
                prev = indices[(i - 1 + m) % m]
                
                # Next index (circular)
                next_ = indices[(i + 1) % m]
                
                # Distance to previous
                d1 = abs(curr - prev)
                d1 = min(d1, n - d1)
                
                # Distance to next
                d2 = abs(curr - next_)
                d2 = min(d2, n - d2)
                
                # Store minimum
                ans[curr] = min(d1, d2)
        
        # Step 4: Answer queries
        return [ans[q] for q in queries]
```
